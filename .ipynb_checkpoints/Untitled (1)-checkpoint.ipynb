# Prétraitement des données

# Mise en échelle

```python
from sklearn.preprocessing import MinMaxScaler
data = df.values
datanorm = MinMaxScaler().fit_transform(data)
datanorm[:5]
```

```python
# Separate features and labels
X=df.iloc[:,:-1].values
y=df.iloc[:, -1].values
```

# Séparation des données en jeu d'entraînement et jeu de test

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=25,stratify=y)
```

```python
X_train.shape
```

```python
y_train = y_train.reshape(y_train.shape[0], 1)
y_train.shape
```

```python
#standarisation des features test et train
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Régression logistique

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(X_train,y_train)
y_test_pred = model1.predict(X_test)
print('Train score : ',model1.score(X_train, y_train))
print('Test score : ',model1.score(X_test, y_test))
```

```python
y_test_pred
```

```python
print("Accuracy = %.f%% des predictions  du model sur le jeu de test sont corectes"
     % (100*metrics.accuracy_score(y_test,y_test_pred)))
metrics.ConfusionMatrixDisplay.from_estimator(model1,X_test, y_test)
plt.show()
```

# Performance du modèle par GridSearchCV

```python

from sklearn.model_selection import GridSearchCV
parameters = {
    'penalty' : ['l1','l2'], 
    'C'       : np.logspace(-3,3,7),
    'solver'  : ['newton-cg', 'lbfgs', 'liblinear'],
}
logreg = LogisticRegression()
clf = GridSearchCV(logreg,                    # model
                   param_grid = parameters,   # hyperparameters
                   scoring='accuracy',        # metric for scoring
                   cv=10)                     # number of folds


logreg_cv = clf.fit(X_train,y_train) 
```

```python
#test logistic regression
print("Tuned hpyerparameters (best parameters):", logreg_cv.best_params_)
print("Train accuracy:", logreg_cv.best_score_)
print("Test accuracy:", logreg_cv.best_estimator_.score(X_test, y_test))
```

# k – Plus proches voisins

```python
from sklearn import neighbors
model2= neighbors.KNeighborsClassifier(n_neighbors=5)
model2.fit(X_train,y_train)
y_test_pred=model2.predict(X_test)
print('Train score : ',model2.score(X_train, y_train))
print('Test score : ',model2.score(X_test, y_test))
```

```python
y_test_pred
```

```python
print("Accuracy = %.f%% des predictions  du model sur le jeu de test sont corectes"
     % (100*metrics.accuracy_score(y_test,y_test_pred)))
metrics.ConfusionMatrixDisplay.from_estimator(model2,X_test, y_test)
plt.show()
```

```python

#train k-nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
parameters = {'n_neighbors': list(range(1, 20)),
              'weights': ['uniform', 'distance'],
              'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
              'p': [1,2]}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, parameters, cv=10, verbose=0)
knn_cv = grid_search.fit(X_train, y_train)
#test k-nearest neighbors
print("Tuned hpyerparameters (best parameters):", knn_cv.best_params_)
print("Train accuracy:", knn_cv.best_score_)
print("Test accuracy:", knn_cv.best_estimator_.score(X_test, y_test))
```

# Performance du modèle par GridSearchCV

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#train artificial neural network
from sklearn.neural_network import MLPClassifier
parameters = {'activation': ['identity', 'logistic', 'tanh', 'relu'],
              'solver': ['lbfgs', 'sgd', 'adam'],
              'alpha': np.logspace(-5, 0, 10),
              'learning_rate': ['adaptive'],
              'max_iter': [1000]}
ann = MLPClassifier()
grid_search = GridSearchCV(ann, parameters, cv=10, verbose=0)
ann_cv = grid_search.fit(X_train, y_train)
#test artificial neural network
print("Tuned hpyerparameters (best parameters):", ann_cv.best_params_)
print("Train accuracy:", ann_cv.best_score_)
print("Test accuracy:", ann_cv.best_estimator_.score(X_test, y_test))
```

# comparaison des modèles

Enfin, nous comparerons directement les précisions des modèles et créerons des matrices de confusion

```python
#plot confusion matrices using model predictions
from sklearn.metrics import confusion_matrix

logreg_cm = confusion_matrix(y_test, logreg_cv.predict(X_test))
knn_cm = confusion_matrix(y_test, knn_cv.predict(X_test))
ann_cm = confusion_matrix(y_test, ann_cv.predict(X_test))
#tree_cm = confusion_matrix(y_test, tree_cv.predict(X_test))
#forest_cm = confusion_matrix(y_test, forest_cv.predict(X_test))
#svm_cm = confusion_matrix(y_test, svm_cv.predict(X_test))
#ann_cm = confusion_matrix(y_test, ann_cv.predict(X_test))
vmax = max(np.amax(logreg_cm), np.amax(knn_cm), np.amax(ann_cm))
fig, axs = plt.subplots(3, constrained_layout=True)
sns.heatmap(logreg_cm, annot=True, cbar=False, ax=axs[0], cmap='mako', fmt='g', vmin=0, vmax=vmax)
sns.heatmap(knn_cm, annot=True, cbar=False, ax=axs[1], cmap='mako', fmt='g', vmin=0, vmax=vmax)
sns.heatmap(ann_cm, annot=True, cbar=False, ax=axs[2], cmap='mako', fmt='g', vmin=0, vmax=vmax)
axs[0].set_title('Logistic Regression')
axs[1].set_title('K-Nearest Neighbors')
axs[2].set_title('Artificial Neural Network')
axs[0].xaxis.set_ticklabels(['0', '1'])
axs[1].xaxis.set_ticklabels(['0', '1'])
axs[2].yaxis.set_ticklabels(['0', '1'])
fig.supxlabel('Predicted label')
fig.supylabel('True label')
plt.show()
```

```python
#compare models
print("Logistic Regression: {}\nKNN: {}\nANN: {}".format(
    logreg_cv.best_estimator_.score(X_test, y_test),
    knn_cv.best_estimator_.score(X_test, y_test),
    ann_cv.best_estimator_.score(X_test, y_test)))
```

Chacun des modèles était presque identique, mais on peut voir que le réseaux de neurone donne les meilleurs résultats.
